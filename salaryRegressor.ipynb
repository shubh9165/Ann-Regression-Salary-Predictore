```python
import tensorflow as tf
import pandas as pd 
import numpy as np
import pickle
from sklearn.preprocessing import StandardScaler,OneHotEncoder,LabelEncoder

```

```python
data=pd.read_csv('Churn_Modelling.csv')
```

```python
data=data.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
```

```python
data
```

```python
label_encoder_gender=LabelEncoder()
data['Gender']=label_encoder_gender.fit_transform(data['Gender'])
```

```python
data
```

```python
one_hot_encoder_geo=OneHotEncoder()
encoded_geo=one_hot_encoder_geo.fit_transform(data[['Geography']]).toarray()
encoded_geo_df=pd.DataFrame(encoded_geo,columns=one_hot_encoder_geo.get_feature_names_out(['Geography']))
```

```python
new_data=pd.concat([data.drop(['Geography'],axis=1),encoded_geo_df],axis=1)
```

```python
new_data
```

```python
X=new_data.drop(['EstimatedSalary'],axis=1)
y=new_data['EstimatedSalary']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.20)
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
with open('StandardScaler.pkl','wb') as file:
    pickle.dump(scaler,file)

with open('OneHotEncoder.pkl','wb') as file:
    pickle.dump(one_hot_encoder_geo,file)

with open('LabelEncoder.pkl','wb') as file:
    pickle.dump(label_encoder_gender,file)
```

# Ann Regression

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential([
    Dense(64,activation='relu',input_shape=(X_train.shape[1],)),
    Dense(32,activation='relu'),
    Dense(1)
])
```

```python
model.compile(optimizer='adam',loss='mean_absolute_error',metrics=['mae'])
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime

log_dir="regressionlogs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
early_stopping_callback = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
history=model.fit(
    X_train,y_train,
    validation_data=[X_test,y_test],
    callbacks=[early_stopping_callback,tensorboard_callback],
    epochs=100
)
```

```python
model.save('regresssionModel.h5')
```

```python
test_loss,test_mae=model.evaluate(X_test,y_test)
```
